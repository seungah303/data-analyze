# 데이터의 부분집합 선책하기

```python
import pandas as pd
df=pd.read_csv('earthquakes.csv')
```

```python
df.mag #mag열을 선택하여 시리즈 객체로 진도데이터 얻기 (딕셔너리)
```

```python
df['mag'] # 속성표기법을 사용하면 딕셔너리와 같은 표기법으로 열 선택 가능
```

```python
df[['mag', 'title']] #한번에 여러 열 선택 가능
```

```python
df[
    ['title', 'time']
    +[col for col in df.columns if col.startswith('mag')]
]
#우리가 요청한 순서로 출력됨.
```
