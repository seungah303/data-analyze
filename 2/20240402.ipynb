# 필터링

```python
df.mag>2
```

```python
df[df.mag>=7.0] #부울 마스크로 필터링한 결과 2개 행만 있는 df얻음
```

```python
df.loc[
    df.mag>=7.0,
    ['alert','mag','magType','title','tsunami','type']
]
#지정한 열만 있는 df 얻을 수 있음. 부울 마스크로 인덱싱 한 결과
```

```python
df.loc[(df.alert=='red') & (df.tsunami==1),['alert','mag','magType','title','tsunami','type']]
#논리곱 비트 연산자로 필터를 결합한 결과
```

```python
df.loc[
    (df.alert=='red') | (df.tsunami==1),
    ['alert','mag','magType','title','tsunami','type']
]
#쓰나미와 적색경보 둘 중 하나만 만족하는 데이터 찾으려면?
```

alert열에 대해서 null이 아닌 알래스카의 모든 지진을 선택해보자!

```python
df.loc[
    (df.place.str.contains('Alaska'))
    & (df.alert.notnull()),
    ['alert','mag','magType','title','tsunami','type']
]
#수치형 데이터가 아닌 열에 부울 마스크를 적용한 결과
```

정규 표현식: 우리가 찾으려는 정확한 내용이 아니라 검색 패턴을 정의할 수 있어 매우 강력!!\
검색 대상이 되는 모든 단어나 숫자를 알지 못해도, 문자열에 있는 모든 단어나 숫자를 찾을 수 있음!\
따옴표 앞에 r 붙이기(raw string 원시문자열 표시임)

```python
df.loc[
    (df.place.str.contains(r'CA|California$'))
    &(df.mag>3.8),
    ['alert','mag','magType','title','tsunami','type']
]
#정규표현식을 사용해 캘리포니아에서 진도가 3.8보다 큰지진을 필터링한 결과
#데이터가 일관성이 없으므로 place 열에서 CA나 California로 끝나는 항목 선택해야 함
```

진도가 6.5에서 7.5 사이인 모든 지진을 찾으려면?
-> 2개의 부울마스크를 쓰면 됨! (하나는 진도가 6.5 이상인지 확인하고, 다른 하나는 7.5 이하인지 확인하는거지~ 그런 다음에 &연산자를 사용해서 두 조건을 결합하면 되겠지!)

하지만 고맙게도!! pandas에서는 between() 메서드를 제공해 이런 마스크를 쉽게 만들수 있어~

```python
df.loc[
    df.mag.between(6.5,7.5),
    ['alert','mag','magType','title','tsunami','type']
]
#값의 범위를 사용해 필터링 한 결과
    
```
