# 정규표현식 : 특정한 조건의 문자를 검색하거나 치환하는 과정을 매우 간편하게 처리하기 위한 수단!!
str.extract()는 괄호 안의 내용들을 추출하겠다는 말~~ unique() 데이터 고유 값을 알고싶을 때 사용 가능

```python
df.place.str.extract(r',(.*$)')[0].sort_values().unique()
```

```python
df['parsed_place']=df.place.str.replace(
 r'.*of', '', regex=True #이라고 하는 모든것을 제거
).str.replace(
    'the', '' #the 제거
).str.replace(
    r'CA$', 'California', regex=True #caliortnia 수정
).str.replace(
    r'NV$', 'Nevada' , regex=True #Nevada 수정
).str.replace(
    r'MX$', 'Mexico', regex=True #Mexico 수정
).str.replace(
    r'region$', '', regex=True #region으로 끝나는 부분 제거
).str.replace(
    'northern', '' #northern 제기
).str.replace(
    'Fiji Islands', 'Fiji', #Fiji로 수정
).str.replace(
    r'^.*,','', regex=True
).str.strip()
```

```python
df.parsed_place.sort_values().unique()
```

```python
import pandas as pd
df=pd.DataFrame({"메뉴":['간장게장','양념게장','토마토게장','스윙스','피자','돈까스','라면','돈까스'], 
                 "가격":['10000원','12000원','7000원','2000원','11000원','7000원','5000원','7000원']})
```

```python
df['가격']=df['가격'].str.replace('원', '')
df
```

```python
df['가격']=pd.to_numeric(df['가격'])
df
```

```python
df.assign(
    in_ca=df.parsed_place.str.endswith('California'),
    in_alaska=df.parsed_place.str.endswith('Alaska')
).sample(5, random_state=0) #임의로 5개 행 선택
```

```python
df.assign(
    in_ca=df.parsed_place=='California',
    in_alaska=df.parsed_place=='Alaska',
    neither=lambda x: ~x.in_ca & ~x.in_alaska
).sample(5, random_state=0) #람다 함수는 많은 기능과 함께 사용할 수 있고, 코드의 품질과 가독성도 획기적으로 향상시킬 수 있음
```
